## 1: Scanner Lookup (Linear Search)

```python
def find_item_linear(arr, target):

    location = -1
    for i in range(len(arr)):

        if arr[i] == target:
            location = i
            break
        
    return location
```

```python
find_item_linear(arr=[5, 5, 5],
                 target=5)
```

## 2: Database Lookup (Binary Search)

```python
def find_item_binary(arr, target):

    location = -1
    left, right = 0, len(arr) - 1

    while left <= right:

        mid = (left + right) // 2

        if arr[mid] == target:

            location = mid
            break

        elif arr[mid] > target:

            right = mid - 1
        
        else:

            left = mid + 1
    
    return location
```

```python
find_item_binary(arr=[1, 3, 5, 7, 9],
                 target=7)
```

## 3: Insertion Slot Finder (Search Insert Position)

```python
def find_insert_position(arr, target):
    
    if len(arr) == 0:

        return 0

    if arr[0] > target:

        return 0
    
    elif arr[-1] < target:

        return len(arr)
    
    else:

        for i in range(len(arr) - 1):

            if target == arr[i]:

                return i
                break

            elif target > arr[i] and target < arr[i + 1]:

                return i + 1


```

```python
find_insert_position(arr = [1, 3, 5, 6],
                     target = 5)
```

## 4: Belt Reordering (Bubble Sort)

```python
def bubble_sort(arr):
    """
    Return a NEW list containing the elements of `arr` sorted in ascending order,
    using the BUBBLE SORT algorithm.

    Do not modify the input list `arr` in place.
    Do NOT call `sorted()` or `.sort()`.

    Example:
        bubble_sort([5,1,4,2]) -> [1,2,4,5]
    """
    result = list(arr)
    for i in range(len(result)):
        for j in range(0, len(result) - i - 1):
            if result[j+1] < result[j]:
                result[j], result[j + 1] = result[j+1], result[j]

    return result
```

```python
bubble_sort([5,1,4,2])
```

# 5: Shelf Selection (Selection Sort)

```python
def selection_sort(arr):
    """
    Return a NEW list containing the elements of `arr` sorted in ascending order,
    using the SELECTION SORT algorithm.

    Do not modify the input list `arr` in place.
    Do NOT call `sorted()` or `.sort()`.

    Example:
        selection_sort([64,25,12,22,11]) -> [11,12,22,25,64]
    """
    new_list = list(arr)
    for i in range(len(arr)):
        min_id = i

        for j in range(i, len(arr)):
            if new_list[j] < new_list[min_id]:
                min_id = j
        new_list[i], new_list[min_id] = new_list[min_id], new_list[i]

    return new_list
```

```python
selection_sort([64,25,12,22,11])
```

# 6: Live Lane Sorting (Insertion Sort)

```python
def insertion_sort(arr):
    """
    Return a NEW list containing the elements of `arr` sorted in ascending order,
    using the INSERTION SORT algorithm.

    Do not modify the input list `arr` in place.
    Do NOT call `sorted()` or `.sort()`.

    Example:
        insertion_sort([5,2,4,6,1,3]) -> [1,2,3,4,5,6]
    """
    result = list(arr)  # make a copy

    for i in range(1, len(result)):
        key = result[i]
        j = i - 1

        # shift elements to the right to make space for key
        while j >= 0 and result[j] > key:
            result[j + 1] = result[j]
            j -= 1

        result[j + 1] = key
    
    return result
```

# 7: Bulk Sorting Service (Merge Sort)

```python
def merge_sort(arr):
    """
    Return a NEW list containing the elements of `arr` sorted in ascending order,
    using the MERGE SORT algorithm (O(n log n)).

    Do not modify the input list `arr` in place.
    Do NOT call `sorted()` or `.sort()`.
    """
    # Base case: if the list has 0 or 1 elements, it's already sorted
    if len(arr) <= 1:
        return arr

    # 1. DIVIDE: Split the list into two halves
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    # 2. CONQUER: Merge the two sorted halves into a new list
    merged = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    # Append any remaining elements from either half
    merged.extend(left[i:])
    merged.extend(right[j:])

    return merged
```

# 8: Express Lane Sorting (Quick Sort)

```python
def quick_sort(arr):
    """
    Return a NEW list containing the elements of `arr` sorted in ascending order,
    using the QUICK SORT algorithm (average O(n log n)).

    Do not modify the input list `arr` in place.
    Do NOT call `sorted()` or `.sort()`.
    """
    # Base case: A list of 0 or 1 elements is already sorted
    if len(arr) <= 1:
        return arr
    
    # 1. Choose a pivot (we'll use the middle element for better average performance)
    pivot = arr[len(arr) // 2]
    
    # 2. Partition: Create three lists based on the pivot
    # Elements smaller than the pivot
    left = [x for x in arr if x < pivot]
    
    # Elements equal to the pivot (handles duplicates)
    middle = [x for x in arr if x == pivot]
    
    # Elements larger than the pivot
    right = [x for x in arr if x > pivot]
    
    # 3. Recursively sort the left and right, then combine
    return quick_sort(left) + middle + quick_sort(right)
```

# 9: K Cheapest Parcels (Top-k Smallest)

```python
def top_k_smallest(arr, k):
    """
    Return a NEW list containing the k smallest elements of `arr`
    in ascending order.

    If k is greater than the length of `arr`, return all elements sorted.

    Example:
        top_k_smallest([7,10,4,3,20,15], 3) -> [3,4,7]
    """
    if k >= len(arr):
        return selection_sort(arr)
    
    else:
        return selection_sort(arr)[:k]
```

```python
top_k_smallest([7,10,4,3,20,15], 3)
```

# 10: Quality Check (Is Sorted?)

```python
def is_sorted(arr):
    """
    Return True if `arr` is sorted in non-decreasing order, otherwise False.

    Examples:
        is_sorted([1,2,2,3]) -> True
        is_sorted([3,2,1])   -> False
        is_sorted([])        -> True
    """
    if len(arr) <= 1:
        return True
    else:
        for i in range(len(arr) - 1):
            if arr[i] > arr[i + 1]:
                return False
        return True
```

```python
is_sorted([1,2,2,3])
```

```python
is_sorted([3,2,1])
```

```python
is_sorted([])
```
