## 1: Scanner Lookup (Linear Search)

```python
def find_item_linear(arr, target):

    location = -1
    for i in range(len(arr)):

        if arr[i] == target:
            location = i
            break
        
    return location
```

```python
find_item_linear(arr=[5, 5, 5],
                 target=5)
```

## 2: Database Lookup (Binary Search)

```python
def find_item_binary(arr, target):

    location = -1
    left, right = 0, len(arr) - 1

    while left <= right:

        mid = (left + right) // 2

        if arr[mid] == target:

            location = mid
            break

        elif arr[mid] > target:

            right -= 1
        
        else:

            left += 1
    
    return location
```

```python
find_item_binary(arr=[1, 3, 5, 7, 9],
                 target=7)
```

## 3: Insertion Slot Finder (Search Insert Position)

```python
def find_insert_position(arr, target):
    
    if len(arr) == 0:

        return 0

    if arr[0] > target:

        return 0
    
    elif arr[-1] < target:

        return len(arr)
    
    else:

        for i in range(len(arr) - 1):

            if target == arr[i]:

                return i
                break

            elif target > arr[i] and target < arr[i + 1]:

                return i + 1


```

```python
find_insert_position(arr = [1, 3, 5, 6],
                     target = 5)
```

```python
find_insert_position(arr = [1, 3, 5, 6],
                     target = 2)
```

```python
find_insert_position(arr = [1, 3, 5, 6],
                     target = 7)
```

```python
find_insert_position(arr = [],
                     target = 3)
```
