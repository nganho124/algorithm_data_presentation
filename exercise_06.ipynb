## 6.1: Constant-time Simulation!

```python
def sum_to_n_fast(n):
    '''(int) -> int
    Returns the sum of all integers from 1 to n inclusive.
    '''
    # TODO: add an assertion that n is non-negative with the message
    #       "n must be non-negative"
    assert n >= 0, "n must be non-negative"

    # TODO: replace this loop with a constant-time formula (no loops)
    total = n * (n + 1) / 2

    return total
```

## 6.2: Efficient Duplicate Check

```python
def has_duplicates(values):
    '''(list) -> bool
    Returns True if values contains any duplicate element, otherwise False.
    '''
    # TODO: implement using a single loop and a set
    #       do not use nested loops
    value_unique = set(values)
    if len(value_unique) < len(values):
        return True
    else:
        return False
```

## 6.4: Frequency Table

```python
def build_frequency_table(values):
    '''(list) -> dict
    Returns a dictionary mapping each element in values to its frequency.
    '''
    # TODO: implement using a single loop and a dictionary
    freq = {}
    for i in range(len(values)):
        if values[i] in freq.keys():
            freq[values[i]] += 1
        else:
            freq[values[i]] = 1

    return freq
```

## 6.5: Most Common Element

```python
def most_common(values):
    '''(list) -> object | None
    Returns the element with the highest frequency in values,
    or None if the list is empty.
    '''
    # TODO:
    # - return None if the list is empty
    # - otherwise reuse build_frequency_table(values)
    # - then scan the dictionary once to find the element with the largest count
    if len(values) == 0:
        return None
    else:
        freq = {}
        for i in range(len(values)):
            if values[i] in freq.keys():
                freq[values[i]] += 1
            else:
                freq[values[i]] = 1

            # Find the maximum value
        max_value = max(freq.values())
        
        # Get all keys with that maximum value
        keys_with_max = [k for k, v in freq.items() if v == max_value]
        
        # If multiple keys have the same max value, return the first one
        return keys_with_max[0]
```

## 6.6: Prefix Sums

```python
def prefix_sums(numbers):
    '''(list[int | float]) -> list[int | float]
    Returns a list where prefix[i] is the sum of numbers[0] up to numbers[i].
    '''
    # TODO: implement prefix sums using a running total in one loop
    result = []
    sum = 0
    for i in range(len(numbers)):
        sum += numbers[i]    
        result.append(sum)
    return result
```

## 6.7: Min and Max in One Pass

```python
def min_max(numbers):
    '''(list[int | float]) -> (int | float, int | float)
    Returns a pair (min_value, max_value) for the list numbers.
    '''
    # TODO: add assertion that numbers is not empty with message
    #       "numbers must not be empty"
    assert len(numbers) > 0, "numbers must not be empty"

    # TODO: implement a single-pass min/max without using min() or max()
    min = numbers[0]
    max = numbers[0]
    for i in range(len(numbers)):
        if numbers[i] < min:
            min = numbers[i]
        if numbers[i] > max:
            max = numbers[i]
    return (min, max)
```

## 6.8: Binary Search

```python
def binary_search(sorted_numbers, target):
    '''(list[int | float], int | float) -> int
    Returns the index of target in sorted_numbers, or -1 if not found.
    '''
    # TODO: implement iterative binary search using low and high indices
    left, right = 0, len(sorted_numbers) - 1

    while left <= right:
        mid = (left + right) // 2
        if sorted_numbers[mid] == target:
            return mid
        elif sorted_numbers[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
```

## 6.9: Fast Fibonacci

```python
def fast_fibonacci(n):
    '''(int) -> int
    Returns the n-th Fibonacci number using an efficient algorithm.
    '''
    # TODO: add an assertion that n is a non-negative integer with message
    #       "n must be a non-negative integer"
    assert n >= 0, "n must be a non-negative integer"

    # TODO: implement an iterative Fibonacci algorithm (no naive recursion)
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        list = [0, 1]
        for i in range(2, n + 1):
            num = list[i-1] + list[i-2]
            list.append(num)

        return list[n]
```
