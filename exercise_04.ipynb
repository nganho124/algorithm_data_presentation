# Task 1: The Ledger of Balance

```python
def merge_sorted(nums1: list[int], 
                m: int, 
                nums2: list[int], 
                n: int) -> None:

    if m == 0 and n != 0:
        return nums2
    
    elif m != 0 and n == 0:
        return nums1

    else:
        nums1 = [x for x in nums1 if x != 0]
        nums1 = nums1 + nums2
        nums1.sort()
        return nums1
    
```

```python
merge_sorted(nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3)
```

```python
merge_sorted(nums1 = [1], m = 1, nums2 = [], n = 0)
```

```python
merge_sorted(nums1 = [0], m = 0, nums2 = [1], n = 1)
```

# Task 2: The Purge of the Cursed Gems

```python
def purge_gems(nums: list[int], val: int) -> int:

    n = 0
    for i in range(len(nums)):
        if nums[i] == val:
            n += 1

    nums = [item for item in nums if item != val]

    # new_list = [""] * n

    # nums = nums + new_list

    return len(nums)
```

```python
def rotate_wheel(nums: list[int], k: int) -> None:

    if k == 0 or k % len(nums) == 0:
        return nums
    
    else:
        n = k % len(nums)

        nums = nums[-n:] + nums[:len(nums) - n]
        return nums
```

```python
rotate_wheel(nums = [1,2,3], k = 6)
```

# Task 5: The Flood of Aqualis

```python
def trapped_water(height: list[int]) -> int:
        if not height:
            return 0
        
        left, right = 0, len(height) - 1
        left_max, right_max = height[left], height[right]
        water = 0
        
        while left < right:
            if left_max <= right_max:
                left += 1
                left_max = max(left_max, height[left])
                water += left_max - height[left]
            else:
                right -= 1
                right_max = max(right_max, height[right])
                water += right_max - height[right]
        
        return water
```

```python
trapped_water(height = [0,1,0,2,1,0,1,3,2,1,2,1])
```

# Task 6: The Twin Reservoirs

```python
def max_reservoir(height: list[int]) -> int:

    max = 0
    for i in range(len(height)):

        for j in range(1, len(height)):

            area = abs(j - i) * min (height[i], height[j])

            if max < area:
                max = area
    
    return max
```

```python
max_reservoir(height = [1,8,6,2,5,4,8,3,7])
```

```python
max_reservoir(height= [1, 1])
```

# Task 7: The Triad of Balance

```python
def find_triads(nums: list[int]) -> list[list[int]]:

    nums.sort()
    result = []
    for i in range(len(nums)):
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        
        start = i + 1
        end = len(nums) - 1
        while start < end:
            sum = nums[i] + nums[start] + nums[end]
            if sum < 0:
                start += 1
            elif sum > 0:
                end -= 1
            else:
                result.append([nums[i], nums[start], nums[end]])
                start += 1
                end -= 1
    
    return result
```

```python
find_triads(nums = [-1,0,1,2,-1,-4])
```

```python
find_triads(nums = [0,1,1])
```

# Task 8: The Spiral of Eternity

```python
def spiral_read(matrix: list[list[int]]) -> list[int]:

    if not matrix:
        return []

    top = 0
    bottom = len(matrix) - 1
    left = 0
    right = len(matrix[0]) - 1

    result = []

    while top <= bottom and left <= right:

        for col in range(left, right + 1):
            result.append(matrix[top][col])
        top += 1

        for row in range(top, bottom + 1):
            result.append(matrix[row][right])
        right -= 1

        if top <= bottom:
            for col in range(right, left - 1, -1):
                result.append(matrix[bottom][col])
            bottom -= 1

        if left <= right:
            for row in range(bottom, top - 1, -1):
                result.append(matrix[row][left])
            left += 1
    
    return result
```

```python
spiral_read(matrix = [])
```

# Task 9: The Rune of Shifting Shadows

```python
def shift_shadows(nums: list[int]) -> None:
    n = 0
    for i in range(len(nums)):
        if nums[i] == 0:
            n += 1

    nums = [item for item in nums if item != 0]

    new_list = [0] * n

    nums = nums + new_list

    return nums
```

```python
shift_shadows(nums = [0,1,0,3,12])
```

```python
shift_shadows(nums = [4,2,4,0,0,3,0,5,1,0])
```

# Task 10: The Balance of Prosperity

```python
def max_prosperity(nums: list[int]) -> int:
    current_sum = nums[0]
    max_sum = nums[0]

    for i in range(1, len(nums)):
        current_sum = max(nums[i], current_sum + nums[i])
        max_sum = max(max_sum, current_sum)

    return max_sum
```
