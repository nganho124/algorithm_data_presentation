## 5.1 Factorial Bugs

```python
def factorial(num):
    '''(int) -> (int)
    Calculates the factorial of a number, given by the variable num.
    '''

    # TODO: Check for the errors in the below code and fix them.

    assert num > 0, "num : Negative number is not allowed"

    output = 1

    while num > 0:
        output = output * num
        num = num - 1

    return output
```

## 5.2 Semantic Errors

```python
def calculate_years_to_double(principal, annual_interest_rate):
    '''(int , int) -> int
    Calculate the number of years required for the given principal to exceed double 
    with the provided annual interest rate using compound interest.
    '''

    # TODO: Use an assertion statement to check if principal is not negative 
    #       and annual_interest_rate is within the range (0, 100].

    assert principal > 0, "Principal cannot be negative"
    assert annual_interest_rate > 0 and annual_interest_rate <= 100, "Interest rate must be in (0, 100]"

    target_amount = principal * 2
    amount = principal
    years = 0

    while amount < target_amount:
        amount = amount*(1 + annual_interest_rate/100)
        years += 1

    return years
```

## 5.5 Off-by-One Mystery

```python
def sum_list(numbers):
    '''(list of int/float) -> int/float
    Returns the sum of all elements in numbers.
    '''

    total = 0
    i = 0

    while i < len(numbers):
        total = total + numbers[i]
        i = i + 1

    return total
```

## 5.6 The Mutating List Crime

```python
def filter_non_negative(values):
    '''(list of int/float) -> list of int/float
    Returns a new list containing only the non-negative values from the input.
    '''

    new_list = []

    for v in values:
        if v >= 0:
            new_list.append(v)

    return new_list
```

## 5.9 The Logical Operator Trap

```python
def check_password(password):
    '''(str) -> bool
    Returns True if password is strong enough, otherwise False.
    '''

    # TODO: Assert that password is a string.
    assert type(password) is str, "Password must be a string"

    has_digit = any(ch.isdigit() for ch in password)
    has_upper = any(ch.isupper() for ch in password)

    if len(password) >= 8 and has_digit and has_upper:
        return True
    else:
        return False
```
